```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#In this notebook CSV file is imported by Pandas module. In columns of the file there are different data about terrorist acts.
#Every row contains another terror act sorted by date. In next steps this data will be analysed with pandas and numpy
#and presented with matplotlib.
```

```python
# 1. What correlation is between weapon type and region of the act of terror?

data = pd.read_csv('globalterrorismdb_0617dist.csv', encoding='ISO-8859-1', usecols=[81,9])
data = data[data['weaptype1']!=13]

print('Correlation between region of the act of terror and weapon type used:',data['region'].corr(data['weaptype1']))

#plt.plot(data['region'])
#plt.plot(data['weaptype1'])
#plt.show()
```

```python
# 2. How hijacking changes over years
data2 = pd.read_csv('globalterrorismdb_0617dist.csv', encoding='ISO-8859-1', usecols=[1,28])
data2 = (data2[data2['attacktype1']==4])
#print(data2)
years, count = np.unique(data2['iyear'],return_counts=True)
#print(years)
#print(count)
plt.plot(years, count)
plt.ylabel('Number of hijackings')
plt.xlabel('Years')
plt.title('How number of hijackings changes over the years')
plt.show()
```

```python
# 3. Difference between number of kidnapped people and the rescued ones

data3 = pd.read_csv('globalterrorismdb_0617dist.csv', encoding='ISO-8859-1', usecols=[1,28,110,124])
data3 = data3[data3['attacktype1']==6]
data3 = data3[data3['nreleased']!=-99]
data3 = data3[data3['nhostkid']!=-99]
data3 = data3[~np.isnan(data3['nreleased'])]

years = np.unique(data3['iyear'])
kidnapped = np.array([])
released = np.array([])
for year in years:
    statInYear = data3[data3['iyear']==year]
    kidnapped = np.append(kidnapped, np.sum(statInYear['nhostkid']))
    released = np.append(released, np.sum(statInYear['nreleased']))
kidnapped1 = np.add.reduceat(kidnapped, np.arange(0, len(kidnapped), 5))
released1 = np.add.reduceat(released, np.arange(0, len(released), 5))
years1 = np.arange(years[0],years[-1],5)
#print(data3)
rects1 = plt.bar(years1, kidnapped1, width=2.5, align='center', label='kidnapped')
rects2 = plt.bar(years1+2.5, released1, width=2.5, align='center', label='released')
plt.tight_layout()
plt.title('Difference between number of kidnapped people and the rescued ones')
plt.xlabel('Years')
plt.ylabel('Number of people')
plt.legend()
plt.show()
```

```python
# 4. Is success dependent on number of preparators?
data4 = pd.read_csv('globalterrorismdb_0617dist.csv', encoding='ISO-8859-1', usecols=[26,69])

data4 = data4[data4['nperps']!=-99]
data4 = data4[data4['nperps']!=0]
data4 = data4[~np.isnan(data4['nperps'])]

print('Correlation between number of terrorists and their success in act of terror:',data4['nperps'].corr(data4['success']))
```

```python
# 5. Percentage of use of different kind of weapons
data5 = pd.read_csv('globalterrorismdb_0617dist.csv', encoding='ISO-8859-1', usecols=[81,])

data5 = data5[data5['weaptype1']!=13]

weapons,count = np.unique(data5['weaptype1'], return_counts=True)

total = np.sum(count)
percent = (count*100)/total
data5 = np.vstack((count, percent))
data5 = pd.DataFrame(data5, ['Total number', 'Percentage'], ['Biological', 'Chemical', 'Radiological', 'Firearms', 'Explosives', 'Fake Weapon', 'Incendiary', 'Melee', 'Vehicle', 'Sabotage Equipment', 'Other'])
print(data5)
```
